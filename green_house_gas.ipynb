# Bar Chart – Top 10 Industries with Highest Emissions

```python
# Sort by emissions and get top 10
top10 = df_clean.sort_values(by='Supply Chain Emission Factors with Margins', ascending=False).head(10)

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(
    x='Supply Chain Emission Factors with Margins',
    y=top10['Industry Code'],
    data=top10,
    palette='Reds_r'
)
plt.title('Top 10 Industries by Emissions')
plt.xlabel('Emissions (kg CO2e per $)')
plt.ylabel('Industry Code')
plt.tight_layout()
plt.show()
```
